```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.style.use('fivethirtyeight')
```

```python
CREDIT_CARD_FRAUD_PATH = './creditcard.csv'
```

```python
credit_card_df = pd.read_csv(CREDIT_CARD_FRAUD_PATH)
credit_card_df
```

```python
credit_cards_class_vc = credit_card_df.Class.value_counts()
credit_cards_class_vc.plot(kind='bar')
plt.show()
```

```python
credit_cards_class_vc.plot(kind='bar', logy=True)
plt.show()
```

```python
V1_class_1 = credit_card_df.V1[credit_card_df['Class'] == 1]
V1_class_1.reset_index(drop=True)
```

```python
V1_class_0 = credit_card_df.V1[credit_card_df['Class'] == 0]
V1_class_0.reset_index(drop=True)
```

```python
?plt.hist
```

```python
plt.hist(V1_class_0, bins=20, orientation="horizontal", density=True, color='grey', alpha = .5)
plt.hist(V1_class_1, bins=20, orientation="horizontal", density=True, color='red', alpha = .5)

plt.legend(['Class 1', 'Class 2'])
plt.xlabel('V1')
plt.show()
```
